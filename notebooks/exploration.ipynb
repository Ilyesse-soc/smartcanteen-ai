```python
from src.config import PATHS
from src.data_generation import generate_synthetic_canteen_dataset

import pandas as pd

PATHS.raw_data_path.parent.mkdir(parents=True, exist_ok=True)

# (Re)génère des données
df = generate_synthetic_canteen_dataset(n_days=365)
df.to_csv(PATHS.raw_data_path, index=False)
print('CSV:', PATHS.raw_data_path)
df.head()
```

```python
import matplotlib.pyplot as plt

tmp = df.copy()
tmp['date'] = pd.to_datetime(tmp['date'])
tmp = tmp.sort_values('date')

plt.figure(figsize=(10,4))
for cid in sorted(tmp['cantine_id'].unique()):
    s = tmp[tmp['cantine_id']==cid].set_index('date')['nb_repas_consommés']
    plt.plot(s.index, s.values, label=f'cantine {cid}', linewidth=1)
plt.title('Nb repas consommés (synthétique)')
plt.legend()
plt.tight_layout()
plt.show()
```
